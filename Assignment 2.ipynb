```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity
import numpy as np

# Assuming df1 is already defined and contains your data
df1_encoded = pd.get_dummies(df1)

obj1 = df1_encoded.iloc[0].values
obj2 = df1_encoded.iloc[1].values

obj1 = np.array(obj1, dtype=float)
obj2 = np.array(obj2, dtype=float)

# Simple Matching Coefficient (SMC)
smc = 1 - pairwise_distances([obj1], [obj2], metric='hamming')[0][0]
print(f"Simple Matching Coefficient: {smc}")

# Cosine Similarity
cos_sim = cosine_similarity([obj1], [obj2])[0][0]
print(f"Cosine Similarity: {cos_sim}")
```
